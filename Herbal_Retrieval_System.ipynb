# Tahap 3
## Text Preprocessing

Tahapan preprocessing:

1. Case Folding
2. Remove Punctuation
3. Tokenization
4. Stopword Removal

Tujuan preprocessing adalah membersihkan teks agar proses pencarian menjadi lebih akurat.

```python
import re
import nltk

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('indonesian'))

print(stop_words)
```

```python
def preprocessing(text):

    # 1. Case Folding
    text = text.lower()

    # 2. Hapus tanda baca
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # 3. Tokenisasi sederhana
    tokens = text.split()

    # 4. Stopword Removal
    tokens = [word for word in tokens if word not in stop_words]

    return " ".join(tokens)
```

```python
df["Clean_Text"] = df["Khasiat"].apply(preprocessing)

df
```

# Tahap 4
## Representasi Dokumen

Dokumen direpresentasikan menggunakan metode TF-IDF (Term Frequency Inverse Document Frequency).

TF-IDF memberikan bobot yang lebih besar pada kata yang penting dan lebih kecil pada kata yang sering muncul.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Membuat objek TF-IDF
vectorizer = TfidfVectorizer()

# Mengubah dokumen menjadi matriks TF-IDF
tfidf_matrix = vectorizer.fit_transform(df["Clean_Text"])

print("TF-IDF berhasil dibuat!")
```

```python
feature_names = vectorizer.get_feature_names_out()

print(feature_names)
```

```python
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names,
    index=df["Tanaman"]
)

tfidf_df
```

# Tahap 5
## Information Retrieval

Model yang digunakan adalah:

Vector Space Model

Kesamaan antara query dan dokumen dihitung menggunakan Cosine Similarity.

```python
print(tfidf_matrix.shape)
```

```python
def search(query):

    # Preprocessing query
    query = preprocessing(query)

    # Ubah query menjadi TF-IDF
    query_vector = vectorizer.transform([query])

    # Hitung cosine similarity
    similarity = cosine_similarity(query_vector, tfidf_matrix)

    return similarity
```

```python
query = "masuk angin"

similarity = search(query)

print(similarity)
```

```python
result = pd.DataFrame({
    "Tanaman": df["Tanaman"],
    "Similarity": similarity.flatten()
})

result
```

```python
result = result.sort_values(
    by="Similarity",
    ascending=False
)

result
```
